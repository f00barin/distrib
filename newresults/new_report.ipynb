# Experiments:

## Setup:

This experiments includes testing with tree different truth (similarity) measures. These are:

* Path Similarity Measure
* UKB Cos Similarity Measure
* UKB Dot Similarity Measure

The 'hat' here corresponds to the term frequently used in journals related to vector space models and it is the SVD of the inner product. 

***In this case, the Matrix hat includes the test terms in the candidate matrix.***





    
    

## PATH SIMILARITY MEASURE - RESULTS

### Setup 1:

Number of Training candidates = 10,000 ; training candidates $\subset$ scoring candidates.

Number of scoring candidates = 25,446 

Number of Testing candidates = 1000 

For Matrix hat:

Number of scoring candidates = 26,446

*Graph 1*: Plot of Average rank vs k - truncation constraint for SVD of the bilinear function matrix

Please note: 

* Basic Training = Using the the projection function.
* hat with no truth matrix = the matrix hat, i.e., SVD(Inner product with all the candidates) 



```python
Image(filename='newresults/path/1000/graphs.png') 
```

*Result 1*: Comparison between: path similarity(source1) set ***vs*** similarity score using the best matrix trained with projection function(source2) ***vs*** similarity score using the best hat matrix(source3)

Top 3 words. 

```python
import extractscript as ex
import samples 

truth, values = ex.exttruthtest('representations/truth-path.hdf5', 'newresults/path/1000/values-used.hdf5')
hat = ex.extractmat('newresults/path/1000/testing-hat-best.hdf5')

test = ex.extractmat('newresults/path/1000/testing-best.hdf5')

dict1 = samples.samplespl(truth, values, 3)
dict2 = samples.samplespl(test, values, 3)
dict3 = samples.samplespl(hat, values, 3)

samples.pprint(dict1, dict2, dict3)
```
